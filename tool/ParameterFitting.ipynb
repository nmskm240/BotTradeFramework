### パラメータ設定

```csharp
var timeframes = Enum.GetValues<Timeframe>();
var param1 = Enumerable.Range(1, 2).Select(x => (decimal)(x * 10));
var param2 = Enumerable.Range(0, 2).Select(x => (decimal)(x * 5 + 15));
var place = ExchangePlace.Bybit;
var symbol = Symbol.Future_BTCUSDT;
var kind = StrategyKind.MACross;
var factory = new BacktestBotFactory();
var request = new ParameterFittingRequest(timeframes, param1, param2);
var usecase = new ParameterFittingUsecase(factory, place, symbol, kind);
var res = await usecase.Call(request);
res.Display();
```
